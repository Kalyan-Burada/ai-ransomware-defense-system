```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import joblib

# 1. LOAD DATASET
# Based on your notebook: Col 0 = ID, Col 1 = Label (1: Ransom, 0: Good), Col 2 = Family
data = pd.read_csv('RansomwareData.csv', header=None)

# 2. SEPARATE FEATURES AND LABELS
y = data.iloc[:, 1]  # Label

X = data.iloc[:, 3:]

X = X.loc[:, (X != 0).any(axis=0)]
print(f"Features remaining after cleaning: {X.shape[1]}")


X_train = X[y == 0] 

# 5. FEATURE SCALING
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# 6. TRAIN THE MODEL (Isolation Forest)
# This model learns the "boundary" of normal behavior
model = IsolationForest(contamination=0.01, random_state=42)
model.fit(X_train_scaled)

# 7. SAVE THE MODEL & SCALER (Module 6)
joblib.dump(model, 'ransomware_defense_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
print("Model trained and saved as 'ransomware_defense_model.pkl'")
```
