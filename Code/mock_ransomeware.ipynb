```python
"""
MOCK RANSOMWARE ATTACK SIMULATOR (Fixed)
=========================================
Simulates ransomware behavior:
  Phase 1 - Mass file creation (high I/O)
  Phase 2 - Rapid rename with extension change (.enc_encrypted_locked)

FIX APPLIED: Added a small sleep between renames so watchdog
can actually catch and batch the events within its 2-second window.
Too fast = events get missed. Controlled speed = reliable detection.
"""

import os
import time
import random
import string

# ── Target the monitored Downloads folder ──────────────────────────────────────
TARGET_DIR = os.path.expanduser("~") + "/Downloads/ATTACK_SIMULATION"
if not os.path.exists(TARGET_DIR):
    os.makedirs(TARGET_DIR)

FILE_COUNT = 200  # Number of files to create and rename


def heavy_io_burst():
    # ── PHASE 1: Mass File Generation ─────────────────────────────────────────
    print(f"\n[!] PHASE 1: Mass File Generation ({FILE_COUNT} files)...")
    for i in range(FILE_COUNT):
        file_path = os.path.join(TARGET_DIR, f"project_data_{i}.docx")
        with open(file_path, "w") as f:
            # Random high-entropy content mimics encrypted data
            f.write(''.join(random.choices(
                string.ascii_letters + string.digits, k=5000
            )))

    print(f"[✓] Phase 1 complete. {FILE_COUNT} files written.\n")
    time.sleep(1)  # Brief pause so monitor is ready before Phase 2

    # ── PHASE 2: Rapid Rename (Simulated Encryption) ───────────────────────────
    print("[!] PHASE 2: Simulating Rapid Encryption (rename burst)...")
    files = os.listdir(TARGET_DIR)

    for filename in files:
        old_path = os.path.join(TARGET_DIR, filename)

        # Only rename original .docx files (skip already-renamed ones)
        if not filename.endswith(".enc_encrypted_locked"):
            new_path = old_path + ".enc_encrypted_locked"

            # os.rename → triggers on_moved in watchdog monitor
            os.rename(old_path, new_path)
            print(f"  Targeting: {filename}")

            # ── KEY FIX ──────────────────────────────────────────────────────
            # Small delay so watchdog can register each event.
            # Without this, renames happen so fast the OS buffers them
            # and watchdog sees them all at once AFTER the burst window closes.
            # 0.05s = 20 renames/sec → well above the >10/2sec threshold.
            time.sleep(0.05)

    print("\n[✓] Attack Cycle Finished. Monitor should have triggered ALERT(s).")


# ── Entry Point ────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    print("=" * 60)
    print("!!! WARNING: STARTING HARSH BEHAVIORAL ATTACK SIMULATION !!!")
    print("=" * 60)
    print("[*] Starting in 3 seconds... switch to your monitor window.")
    time.sleep(3)
    heavy_io_burst()
```

```python
"""
MOCK RANSOMWARE ATTACK SIMULATOR — FINAL VERSION
=================================================
Simulates ransomware behavior in ~/Downloads/ATTACK_SIMULATION:
  Phase 1 — Mass file creation  (200 .docx files)
  Phase 2 — Rapid rename burst  (adds .enc_encrypted_locked extension)

The 0.05s delay between renames keeps event rate at ~20/sec,
well above the monitor's >5/2s detection threshold.
"""

import os
import time
import random
import string

TARGET_DIR = os.path.expanduser("~") + "/Downloads/ATTACK_SIMULATION"
os.makedirs(TARGET_DIR, exist_ok=True)

FILE_COUNT   = 200
RENAME_DELAY = 0.05   # seconds between renames — controls detection reliability


def heavy_io_burst():
    # ── Phase 1: Mass File Creation ───────────────────────────────────────────
    print(f"\n[!] PHASE 1: Mass File Generation ({FILE_COUNT} files)...")
    for i in range(FILE_COUNT):
        fpath = os.path.join(TARGET_DIR, f"project_data_{i}.docx")
        with open(fpath, "w") as f:
            f.write(''.join(random.choices(
                string.ascii_letters + string.digits, k=5000
            )))
    print(f"[✓] Phase 1 complete. {FILE_COUNT} files written.")

    time.sleep(1)   # Let monitor settle before Phase 2

    # ── Phase 2: Rapid Rename (Encryption Simulation) ─────────────────────────
    print("\n[!] PHASE 2: Simulating Rapid Encryption (rename burst)...")
    files = [f for f in os.listdir(TARGET_DIR) if not f.endswith(".enc_encrypted_locked")]

    for filename in files:
        old_path = os.path.join(TARGET_DIR, filename)
        new_path = old_path + ".enc_encrypted_locked"
        os.rename(old_path, new_path)
        print(f"  Targeting: {filename}")
        time.sleep(RENAME_DELAY)   # Controlled rate so watchdog catches events

    print("\n[✓] Attack Cycle Finished. Monitor should have triggered ALERT(s).")


if __name__ == "__main__":
    print("=" * 58)
    print("  !!! HARSH BEHAVIORAL ATTACK SIMULATION STARTING !!!")
    print("=" * 58)
    print("[*] Starting in 3 seconds — switch to your monitor window.")
    time.sleep(3)
    heavy_io_burst()
```
